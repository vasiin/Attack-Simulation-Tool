```python

import json

def find_high_threshold_entries(data, threshold):
    high_threshold_entries = []

    for entry in data:
        if entry["response_time_ms"] > threshold:
            high_threshold_entries.append(entry)

    return high_threshold_entries

# Read JSON data from file
file_path = r'C:\Users\User\Desktop\Log_DDoS.txt'
with open(file_path, 'r') as file:
    json_text = file.read()

# Print loaded JSON data for debugging
print("Loaded JSON data:")
print(json_text)

# Example usage
json_data = json.loads(json_text)

# Print response times for debugging
response_times = [entry["response_time_ms"] for entry in json_data]
print("Response times in the data:")
print(response_times)

threshold = 15  # Replace with your desired threshold
high_threshold_entries = find_high_threshold_entries(json_data, threshold)

print("Entries with response time higher than {} ms:".format(threshold))
for entry in high_threshold_entries:
    print(entry)
```
