```python
import json

def detect_xss_entries(json_data):
    xss_entries = set()
    count = 0

    def is_potential_xss(value):
        # Check for potential XSS patterns
        xss_patterns = ['<script>', 'javascript:', 'onerror=', 'alert(']
        for pattern in xss_patterns:
            if pattern in value:
                return True
        return False

    for entry in json_data:
        has_xss = False
        for key, value in entry.items():
            if isinstance(value, str) and is_potential_xss(value):
                has_xss = True
                break

        # Check URL parameters for potential XSS
        if "url" in entry and isinstance(entry["url"], str) and is_potential_xss(entry["url"]):
            has_xss = True

        if has_xss:
            xss_entries.add(json.dumps(entry))
            count += 1

    return count, [json.loads(entry) for entry in xss_entries]

file_path = r"C:\Users\User\Desktop\log simulation\LogXSS.txt"
try:
    with open(file_path, "r") as file:
        log_data = json.load(file)
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")

# Detect entries related to XSS attacks
count, xss_entries = detect_xss_entries(log_data)

# Output the result
print(f"Number of potential XSS entries: {count}")
print(f"Entries related to potential XSS attacks:")
for entry in xss_entries:
    print(entry)

# เขียนข้อมูลที่จำแนกแล้วลงไปในไฟล์ JSON 
# Detect entries related to XSS attacks
count, xss_entries = detect_xss_entries(log_data)

# Output the result
print(f"Number of potential XSS entries: {count}")
print(f"Entries related to potential XSS attacks:")
for entry in xss_entries:
    print(entry)

# Save the detected entries to a JSON file
xss_file_path = r"C:\Users\User\Desktop\log simulation\DetectedXSS.json"
try:
    with open(xss_file_path, "w") as outfile:
        json.dump(xss_entries, outfile)
    print(f"Detected XSS entries saved to '{xss_file_path}'")
except IOError as e:
    print(f"Error writing to file: {e}")


```
